```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# load the dataset
data = pd.read_csv('housing_rents.csv')
# display the first few rows of the dataset
print(data.head())
```

```python
df_encoded = pd.get_dummies(data, columns=["location"], drop_first=False)
df_encoded.head(10)
```

```python
# Prepare the data
X = df_encoded[["size", "location_hyderabad", "location_chennai", "location_bangalore"]]
y = df_encoded["rent"]

# X = data[['size','location']]
# y = data['rent']

# Create and train the model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Evaluation
mse = mean_squared_error(y, y_pred)
print("MSE:", mse)
```

```python
# Results
print("Slope for size:", model.coef_[0])
print("Slope for location:", model.coef_[1])
print("Intercept:", model.intercept_)
```

```python
new_house = pd.DataFrame({"size": [1000], "location": [3]})
predicted_rent = model.predict(new_house)
print("Predicted rent:", predicted_rent[0])
```
