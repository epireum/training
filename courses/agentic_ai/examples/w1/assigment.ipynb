# Pokeman table
1. API Endpoint

Use the Pokémon list endpoint:
https://pokeapi.co/api/v2/pokemon/
This endpoint is paginated, meaning:
It returns only a limited number of Pokémon per request.
It includes a "next" field with the URL for the next page.
Your code must follow pagination until there are no more pages.

2. What to Extract
For each Pokémon, collect:
id|name|height|weight|base_experience|types(comma-separated list)

3. Store data to csv file using pandas
