int: Whole numbers (e.g., 42, -17, 0)
float: Decimal numbers (e.g., 3.14, -0.5, 2.0)
str: Text strings (e.g., "Hello", 'Python')
bool: Boolean values (True or False)

```python
# String variables
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name
print(f"Hello, {full_name} !")
```

```python
# List holds multiple values
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

```python
# Dictionary holds key-value pairs
person = {"name": "Alice", "age": 30, "city": "New York"}
print(person["name"])
```

```python
# function are reusable blocks of code used to perform a specific task
def add_two_number(n1, n2):
    return n1 + n2

result = add_two_number(5, 10)
print(result)
```

```python
# Module is a file containing Python code that can define functions, classes, and variables. A module allows you to organize your code into manageable sections and reuse it across different programs.
import math
print(math.sqrt(16))

```

```python

# Class is a blueprint for creating objects. It defines a set of attributes and methods that the created objects will have.
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        return "Woof!"
my_dog = Dog("Buddy", 3)
print(my_dog.name)
print(my_dog.bark())
```

```python
# JSON (JavaScript Object Notation) is a lightweight data interchange format that is easy for humans to read and write, 
# and easy for machines to parse and generate. It is commonly used for transmitting data in web applications.
import json
data = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}

# Convert Python dictionary to JSON string
json_data = json.dumps(data)
print(json_data)

# Parse JSON string back to Python dictionary
parsed_data = json.loads(json_data)
print(parsed_data["name"])
```

```python
# Calling APIs
#%pip install requests

import requests
response = requests.get("https://pokeapi.co/api/v2/pokemon/pikachu")
if response.status_code == 200:
    data = response.json()
    print(data["name"])
else:
    print("Error:", response.status_code)   
```

```python
# File Handling

# Writing to a file
with open("../temp/data.json", "w") as f:
    json.dump(data, f)
```

```python
# File Handling

# Reading from file
with open("../temp/data.json", "r") as f:
    data = json.load(f)
    print(data["name"])
```
