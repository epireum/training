## Convert PDF file to text format

```python
pdf_file = 'Yellow_Annual_Statement_2024.pdf'
reader = PdfReader(pdf_file)
num_pages = len(reader.pages)
output_text_file = 'extracted_text.txt'
all_text = []
for page_num in range(num_pages):
    page = reader.pages[page_num]
    text = page.extract_text()
    if text and text.strip():
        all_text.append(text)
    else:
        images = convert_from_path(pdf_file, first_page=page_num + 1, last_page=page_num + 1)
        for image in images:
            ocr_text = pytesseract.image_to_string(image)
            all_text.append(ocr_text)
with open(output_text_file, 'w', encoding='utf-8') as f:
    f.write('\n'.join(all_text))

print(f'Text extraction complete. Extracted text saved to {output_text_file}.')    
```
